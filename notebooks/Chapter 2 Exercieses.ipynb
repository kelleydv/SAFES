## Simulating rolls of a fair die

```python
%matplotlib inline

import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from safes.die import Die
```

### Maximum absolute deviation

```python
def mad(n, sides=6):
    d = Die(sides=sides, n=n)
    return (d.frequency - 1/sides).abs().max()
deviations = [mad(x) for x in range(6,2000)] #up to 2000 rolls
```

```python
y1 = np.array(deviations)
plt.plot(y1)
plt.xlabel('number of rolls')
plt.ylabel('deviation')
plt.show()
```

### Average absolute deviation

```python
def aad(n, sides=6):
    d = Die(sides=sides, n=n)
    return (d.frequency - 1/sides).abs().mean()
deviations = [aad(x) for x in range(6,3000)] #up to 3000 rolls
```

```python
y2 = np.array(deviations)
plt.plot(y2)
plt.xlabel('number of rolls')
plt.ylabel('deviation')
plt.show()
```

### How probable would it be to observe the following frequency distribution for rolling a fair 12-sided die 250 times?

```python
obs = pd.Series([20,24,20,19,25,19,19,17,24,20,19,24], index=range(1,13))
print(obs)
```

```python
(obs/250 - 1/12).abs().mean()
```

#### Consider the absolute average deviation from the theoretical frequency distribution above against the plot below

```python
deviations = np.array([aad(x, sides=12) for x in range(12,2000)])
plt.plot(deviations)
plt.xlabel('number of rolls')
plt.ylabel('deviation')
plt.show()
```
